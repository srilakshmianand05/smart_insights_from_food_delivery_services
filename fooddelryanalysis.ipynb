```python
#TOP 10 RESTAURANTS WITH HIGHEST NUMBER OF ORDERS

import pymysql
import pandas as pd
import matplotlib.pyplot as plt

conn = pymysql.connect(
    host="localhost",
    user="root",
    password="srilakshmi1006",
    database="fooddelry_analysis"
)
cursor = conn.cursor()
query = """
select restaurant_id,count(order_id) as total_orders
from order_data
group by restaurant_id
order by total_orders desc
limit 50;
"""
cursor.execute(query)
data = cursor.fetchall()

df = pd.DataFrame(data,columns=["restaurant_id","order_id"])
cursor.close()
conn.close()

top_10 = df.nlargest(10, "order_id")


plt.figure(figsize=(8,5))
plt.barh(top_10["restaurant_id"], top_10["order_id"], color="purple")
plt.xlabel("restaurant_id")
plt.ylabel("order_id")
plt.title("Top 10 restaurants with highest number of orders")
plt.xticks(rotation=45)
plt.show()
```

```python
#TOP RESTAURANTS WITH HIGHEST REVENUE

conn = pymysql.connect(host="localhost", user="root", password="srilakshmi1006", database="fooddelry_analysis")
cursor = conn.cursor()
query = """
select restaurant_id, sum(order_value) as total_revenue
from order_data
group by restaurant_id
order by total_revenue desc
limit 20;
"""
cursor.execute(query)
data = cursor.fetchall()
df = pd.DataFrame(data, columns=["restaurant_id", "total_revenue"])
cursor.close()
conn.close()

plt.figure(figsize=(8,5))
plt.bar(df["restaurant_id"], df["total_revenue"], color="olive")
plt.xlabel("restaurant_id")
plt.ylabel("total_revenue")
plt.title("Top 20 restaurants with highest revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
#HOURLY ORDER TRENDS

import pymysql
import pandas as pd
import matplotlib.pyplot as plt

conn = pymysql.connect(host="localhost", user="root", password="srilakshmi1006", database="fooddelry_analysis")
cursor = conn.cursor()

query = """
select order_id,order_date_and_time from order_data;
"""
cursor.execute(query)
data = cursor.fetchall()

df = pd.DataFrame(data, columns=["order_id","order_date_and_time"])
df["order_date_and_time"] = pd.to_datetime(df["order_date_and_time"])
df["order_hour"] = df["order_date_and_time"].dt.hour
hourly_order_trends = df.groupby("order_hour").size().reset_index(name="total_orders")

cursor.close()
conn.close()

plt.figure(figsize=(8,5))
plt.plot(hourly_order_trends["order_hour"], hourly_order_trends["total_orders"], marker="o", color="red")
plt.xticks(hourly_order_trends["order_hour"])
plt.xlabel("order_hour")
plt.ylabel("total_orders")
plt.title("Hourly order trends")
plt.grid(True,linestyle="--",alpha=0.7)
plt.show()
```

```python
#DAILY ORDER TRENDS

import pymysql
import pandas as pd
import matplotlib.pyplot as plt

conn = pymysql.connect(host="localhost", user="root", password="srilakshmi1006", database="fooddelry_analysis")
cursor = conn.cursor()
query = """
select date(order_date_and_time) as date, count(order_id) as total_orders
from order_data
group by date(order_date_and_time)
order by date;
"""
cursor.execute(query)
data = cursor.fetchall()
df = pd.DataFrame(data, columns=["date", "total_orders"])
cursor.close()
conn.close()

plt.figure(figsize=(8,5))
plt.plot(df["date"], df["total_orders"], marker="o", color="cornflowerblue")
plt.xlabel("date")
plt.ylabel("total_orders")
plt.title("Daily order trends")
plt.grid()
plt.show()
```

```python
#PAYMENT METHOD DISTRIBUTION

conn = pymysql.connect(host="localhost", user="root", password="srilakshmi1006", database="fooddelry_analysis")
cursor = conn.cursor()

query = """
select payment_method, sum(order_value) as total_revenue
from order_data
group by payment_method;
"""

cursor.execute(query)
data = cursor.fetchall()

df = pd.DataFrame(data, columns=["payment_method", "total_revenue"])
cursor.close()
conn.close()

plt.figure(figsize=(8,5))
plt.pie(df["total_revenue"], labels=df["payment_method"], autopct="%1.1f%%", startangle=90, colors=["skyblue","lightcoral","lightgreen"])
plt.axis("equal")
plt.title("Payment method distribution")
plt.show()
```

```python
#TOP 5 RESTAURANTS WITH HIGHEST REFUNDS

conn = pymysql.connect(host="localhost", user="root", password="srilakshmi1006", database="fooddelry_analysis")
cursor = conn.cursor()
query = """
select restaurant_id, sum(refunds) as total_refunds
from order_data
group by restaurant_id
order by total_refunds desc
limit 10;
"""
cursor.execute(query)
data = cursor.fetchall()
df = pd.DataFrame(data, columns=["restaurant_id", "total_refunds"])
cursor.close()
conn.close()

plt.figure(figsize=(8,5))
plt.hist(df["total_refunds"], bins=5, color="orange")
plt.xlabel("restaurant_id")
plt.ylabel("total_refunds")
plt.title("Top 10 restaurants with highest refunds")
plt.show()
```

```python
#DISCOUNTS AND OFFERS DISTRIBUTION

conn = pymysql.connect(host="localhost", user="root", password="srilakshmi1006", database="fooddelry_analysis")
cursor = conn.cursor()

query = """
select discounts_and_offers, sum(order_value) as total_revenue
from order_data
group by discounts_and_offers;
"""
cursor.execute(query)
data = cursor.fetchall()
df = pd.DataFrame(data, columns=["discounts_and_offers", "total_revenue"])
cursor.close()
conn.close()

plt.figure(figsize=(8,5))
plt.pie(df["total_revenue"], labels=df["discounts_and_offers"], autopct="%1.1f%%", startangle=90, colors=["cornsilk","crimson","darkblue","darkseagreen"])
plt.axis("equal")
plt.title("Discounts and offers distribution")
plt.show()
```

```python
#TOP 10 RESTAURANTS WITH HIGHEST ORDER VALUE AND COMMISSION FEE

import pymysql
import pandas as pd
import matplotlib.pyplot as plt

conn = pymysql.connect(host="localhost", user="root", password="srilakshmi1006", database="fooddelry_analysis")
cursor = conn.cursor()

query = """
select restaurant_id, sum(order_value) as total_order_value, sum(commission_fee) as total_commission_fee
from order_data
group by restaurant_id
order by total_order_value desc
limit 10;
"""
cursor.execute(query)
data = cursor.fetchall()
df = pd.DataFrame(data, columns=["restaurant_id", "total_order_value", "total_commission_fee"])
cursor.close()
conn.close()

df.set_index("restaurant_id", inplace=True)
df.plot(kind="bar", stacked=True, figsize=(10,6), color=["skyblue","lightcoral"])
plt.xlabel("restaurant_id")
plt.ylabel("total_order_value")
plt.title("Top 10 restaurants with highest order value and commission fee")
plt.xticks(rotation=45)
plt.show()
```
